```python
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Training Dataset

```python
# load images in the train dataset
train_pneumo = glob('chest_xray/train/PNEUMONIA/*.jpeg')
train_normal = glob('chest_xray/train/NORMAL/*.jpeg')
```

```python
train_pneumo[0]
```

```python
# view a normal image from the train set
pneumo = plt.imread(train_normal[0])

plt.imshow(pneumo, cmap='Greys_r')
print('Normal')
print(pneumo.shape)
```

```python
# view a pneumonia image from the train set
pneumo = plt.imread(train_pneumo[0])

plt.imshow(pneumo, cmap='Greys_r')
print('Pnuemonia')
print(pneumo.shape)
```

```python
train_total = train_normal + train_pneumo
```

```python
# getting a df with normal category
df_normal = pd.DataFrame(['normal']*len(train_normal) , columns = ['category'])

# getting a df with pneumo category
df_pneumo = pd.DataFrame(['pneumonia']*len(train_pneumo) , columns = ['category'])

# concat both df
df_train = pd.concat([df_normal, df_pneumo], ignore_index=True)

# add corresponding images to each category
df_train['image'] = [x for x in train_total]
```

```python
df_train
```

```python
# see the # of the samples
df_train.category.value_counts()
```

```python
sns.countplot(df_train['category'])
plt.title('Number of Training Samples', fontsize = 15)
plt.xlabel('Category', fontsize = 12)
plt.ylabel('Count', fontsize = 12)
plt.xticks([0, 1], ['Normal', 'Pneumonia'], fontsize = 10)

plt.savefig('Training Samples')
```
